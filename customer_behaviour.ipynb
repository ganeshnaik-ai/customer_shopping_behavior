```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
df=head()
```

```python
df.isnull().sum()
```

```python
df['Review Rating']= df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df.columns
```

```python
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
df.columns
```

```python
# create new columns  age_groups
labels = ['young_adults','adults','middle_aged','senoir']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
 # create columns purchase_frequency-days
freq_mapping={
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months':90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(freq_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied']==df['promo_code_used']).all()
```

```python
df=df.drop('promo_code_used',axis=1)
```

```python
df.columns
```

## **Kpi analysis**

### ***Gender***

```python
gender_counts = df['gender'].value_counts()
print(gender_counts)
```

```python
total_revenue = df['purchase_amount'].sum()
print("Total Revenue:", total_revenue)
```

```python
top_item = df['item_purchased'].value_counts().head(10)
print(top_item)
```

```python
location_item = df.groupby('location')['item_purchased'].count().sort_values(ascending=False)
print(location_item.head())
```

```python
age_group = df['age_group'].value_counts()
print(age_group)
```

```python
freq_items = df['frequency_of_purchases'].value_counts()
print(freq_items)
```

```python
category_revenue = df.groupby('category')['purchase_amount'].sum()
print(category_revenue)
```

```python
payment_method = df['payment_method'].value_counts()
print(payment_method)
```

```python
discount_sales = df.groupby('discount_applied')['purchase_amount'].sum()
print(discount_sales)
```

```python
corr = df.corr(numeric_only=True)
print(corr)
```

## ***KPI Analysis***

```python
print("Male vs Female Purchases:",gender_counts)
print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
print("Total Revenue:",total_revenue)
print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
print("Top Most Sold Item:",top_item)
print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
print("Category-wise Revenue :",category_revenue)
print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
print("Location vs Items Sold:",location_item.head())
print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
print("Most Customers by Age Group :",age_group)
print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
print("Frequency of Purchased Items :",freq_items)
print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
print("Discount Impact on Sales :",discount_sales)
```

## **Visualization**

```python
gender_counts.plot(kind='pie')
plt.title("Male vs Female Purchases")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
top_item_revenue = df.groupby('item_purchased')['purchase_amount'].sum().sort_values(ascending=False).head(10)
top_item_revenue.plot(kind='pie')
plt.title("Top 10 Most Sold Items")
plt.xlabel("Item")
plt.ylabel("Count")
plt.show()
```

```python
location_item.head(10).plot(kind='bar')
plt.title("Top Locations by Sales")
plt.xlabel("Location")
plt.ylabel("Items Sold")
plt.show()
```

```python
age_group.plot(kind='bar')
plt.title("Customers by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.show()
```

```python
freq_items.plot(kind='line')
plt.title("Purchase Frequency Distribution")
plt.xlabel("Frequency")
plt.ylabel("Count")
plt.show()
```

```python
category_revenue.plot(kind='line')
plt.title("Revenue by Category")
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.show()
```

```python
payment_method.plot(kind='bar')
plt.title("Payment Method Distribution")
plt.xlabel("Method")
plt.ylabel("Count")
plt.show()
```

```python
discount_sales.plot(kind='bar')
plt.title("Discount Impact on Revenue")
plt.xlabel("Discount Applied")
plt.ylabel("Revenue")
plt.show()
```

```python
plt.imshow(corr)
plt.colorbar()
plt.title("Correlation Heatmap")
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.show()
```
